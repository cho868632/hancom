```python
from ultralytics import solutions
import cv2
from getpass import getpass
```

```python
cam= cv2.VideoCapture(0)
```

```python
stream_url = "https://strm1.spatic.go.kr/live/36.stream/chunklist_w626582203.m3u8"
cam = cv2.VideoCapture(stream_url)
```

```python
from_email = "ach10005302@gmail.com"
password = getpass("앱 비밀번호: ")
to_email = "cho868632@gmail.com"
```

```python
google_alarm = solutions.SecurityAlarm(
    model = "yolo26s.pt",
    show=False,
    classes=[2],
)
```

```python
google_alarm.authenticate(from_email, password, to_email)
```

```python
while cam.isOpened():
    success, frame = cam.read()
    if not success:
        print("비디오 읽기 실패 . . .")
        break

 
    results = google_alarm(frame)
    cv2.imshow("ALARM",results.plot_im)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("q키를 눌러서 종료")
        break
```

```python
cam.release()
cv2.destroyAllWindows()
```
