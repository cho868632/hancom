```python
model = YOLO("yolo26s.pt")
```

```python
model.names
```

```python
model(
    "images/input_params.jpg",

    save=True,
    classes=[67, 75],
    conf=0.5,            # 신뢰도 50% 이상만 탐지
    imgsz=640,           # 추론 이미지 크기
    max_det=3,            # 최대 1개만 탐지
    save_crop=True,       # 탐지 영역 이미지 저장
    save_txt=True,        # 좌표 텍스트 저장
    save_conf=True        # 신뢰도 저장
)
```
