```python
from ultralytics import YOLO
import cv2
```

```python
cap = cv2.VideoCapture("https://strm1.spatic.go.kr/live/36.stream/chunklist_w626582203.m3u8")
```

```python
model = YOLO("yolo26s.pt")
```

```python
while cap.isOpened():
    success,frame = cap.read()
    if not success:
        print("웹캠읽기실패")
        break
    res = model(frame)
    annotated_frame = res[0].plot()

    cv2.imshow("webcam",annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
cap.release()
cv2.destroyAllWindows()
```
