```python
import os
from dotenv import load_dotenv
from huggingface_hub import InferenceClient
load_dotenv()
API_KEY = os.getenv("API_KEY")
```

---

# Kimi-K2.6

---



```python
client = InferenceClient(
    api_key=API_KEY,
)
answer = input("질문을 입력해 주세요: ")
completion = client.chat.completions.create(
    model="moonshotai/Kimi-K2.6:novita",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": answer
                }
                
            ]
        }
    ],
)

print(completion.choices[0].message)
```

---

# Krea-2-Turbo

---

```python
client = InferenceClient(
    provider="fal-ai",
    api_key=API_KEY,
)
answer = input("질문을 입력하세요")

# output is a PIL.Image object
image = client.text_to_image(
    answer,
    model="krea/Krea-2-Turbo",
)
image.save("output.png") 
```

# Inkling

---


```python
import base64


client = InferenceClient(
    api_key=API_KEY,
)

# 로컬 스크린샷 파일을 base64로 인코딩
image_path = "image.png"  # 본인의 UI 스크린샷 경로
with open(image_path, "rb") as f:
    image_base64 = base64.b64encode(f.read()).decode("utf-8")

image_data_url = f"data:image/png;base64,{image_base64}"

completion = client.chat.completions.create(
    model="thinkingmachines/Inkling:together",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": (
                        "이 스크린샷과 최대한 똑같은 화면을 만드는 "
                        "HTML과 CSS 코드를 작성해줘. "
                        "레이아웃, 색상, 폰트, 여백까지 최대한 비슷하게 맞춰줘. "
                        "코드는 하나의 HTML 파일 안에 <style> 태그로 CSS를 포함해서 작성해줘."
                    )
                },
                {
                    "type": "image_url",
                    "image_url": {
                        "url": image_data_url
                    }
                }
            ]
        }
    ],
)

with open("result.html", "w", encoding="utf-8") as f:
    f.write(completion.choices[0].message.content)
```
