```python
from ultralytics import solutions
import cv2
```

```python
cam= cv2.VideoCapture(0)
```

```python
stream_url = "https://strm1.spatic.go.kr/live/36.stream/chunklist_w626582203.m3u8"
cam = cv2.VideoCapture(stream_url)
```

```python
distance = solutions.DistanceCalculation(
    model="yolo26s.pt",
    show=True
)
```

```python
NEAR = 150
MID = 350

def judge(px):
    if px == 0:
        return "WAIT"
    if px < NEAR:
        return "NEAR"
    if px < MID:
        return "MID"
    return "FAR"
```

```python
while cam.isOpened():
    success, frame = cam.read()
    if not success:
        print("비디오 읽기 실패 . . .")
        break

 
    results = distance(frame)
    px= results.pixels_distance
    print(judge(px))
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("q키를 눌러서 종료")
        break
```

```python
cam.release()
cv2.destroyAllWindows()
```
