```python
from ultralytics import YOLO
import cv2

WARNING_THRESHOLD = 5
```

```python
stream_url = "https://strm1.spatic.go.kr/live/36.stream/chunklist_w626582203.m3u8"
cap = cv2.VideoCapture(stream_url)
```

```python
model = YOLO("yolo26s.pt")
```

```python
while cap.isOpened():
    success,frame = cap.read()
    if not success:
        print("웹캠읽기실패")
        break
    res = model(frame)
    annotated_frame = res[0].plot()

    cnt = len(res[0].boxes)
    print(res[0])

    if cnt >= WARNING_THRESHOLD:
        status = "Warning"
        color = (0,0,255)
    else:
        status = "safe"
        color = (255,0,0)
    
    cv2.putText(
        annotated_frame,
        f"Detected : {cnt},{status}",
        (10,30),
        cv2.FONT_HERSHEY_SIMPLEX,1,color,2,cv2.LINE_AA
    )

    cv2.imshow("webcam",annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

## 차/버스만 카운트

`model.names` 로 클래스명 확인 가능. COCO 기준 `2=car`, `5=bus`, `7=truck`.

```python
CAR_ID, BUS_ID = 2, 5   # COCO: 2=car, 5=bus

cap = cv2.VideoCapture(stream_url)

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        print("웹캠읽기실패")
        break

    res = model(frame, classes=[CAR_ID, BUS_ID], verbose=False)
    annotated_frame = res[0].plot()

    cls = res[0].boxes.cls.tolist()
    car_cnt = cls.count(CAR_ID)
    bus_cnt = cls.count(BUS_ID)
    cnt = car_cnt + bus_cnt

    if cnt >= WARNING_THRESHOLD:
        status, color = "Warning", (0, 0, 255)
    else:
        status, color = "safe", (255, 0, 0)

    cv2.putText(
        annotated_frame,
        f"Car:{car_cnt} Bus:{bus_cnt} Total:{cnt} {status}",
        (10, 30),
        cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2, cv2.LINE_AA
    )

    cv2.imshow("webcam", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
cap.release()
cv2.destroyAllWindows()
```
